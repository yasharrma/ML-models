```python
X_train , X_test , y_train , y_test = train_test_split(df.drop(columns = ['median_house_value']),
                                                       df['median_house_value'],
                                                       test_size = 0.2 ,
                                                       random_state = 42 )
```

```python
trf1 = ColumnTransformer([
    ('immpute_bedrooms' ,  SimpleImputer(),[4])]
    , remainder = 'passthrough' )
```

```python
trf2 = ColumnTransformer([
    ('ohe_ocean' , OneHotEncoder(sparse_output = False , handle_unknown = 'ignore'),[8])],
    remainder = 'passthrough' )
```

```python
trf3 = RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rfg_mse = mean_squared_error(y_test,y_pred)
rfg_rmse = np.sqrt(rfg_mse)
rfg_rmse
```

```python
# Define Hyperparameter Grid
param_grid = {
    'trf3__n_estimators': [50, 100, 200],  # Number of trees
    'trf3__max_depth': [None, 10, 20],  # Tree depth
}

# Define Inner Cross-Validation for Grid Search
inner_cv = KFold(n_splits=5, shuffle=True, random_state=42)
grid_search = GridSearchCV(pipe, param_grid, cv=inner_cv, n_jobs=-1, scoring='neg_mean_squared_error')

# Define Outer Cross-Validation for Model Evaluation
outer_cv = KFold(n_splits=5, shuffle=True, random_state=42)
nested_scores = cross_val_score(grid_search, X_train, y_train, cv=outer_cv, scoring='neg_mean_squared_error', n_jobs=-1)

# Convert Negative MSE to Positive RMSE
rmse_scores = np.sqrt(-nested_scores)

# Print Nested CV Results
print("Nested CV RMSE Scores:", rmse_scores)
print("Mean RMSE:", rmse_scores.mean())
```

```python
grid_search.fit(X_train,y_train)
final_model = grid_search.best_estimator_
```

```python
final_predictions = final_model.predict(X_test)

final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
final_rmse
```
